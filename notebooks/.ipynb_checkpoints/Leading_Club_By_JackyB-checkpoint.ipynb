### Null Values Inspection <a name="id5"></a>

Due to the big number of variables, a way to check the nulls values is defining a function to count it and to extract a dataframe the variables with a higher proportion of null values. 

```python
def counting_nullval (df):
    nanval_perct = ((100 * df.isnull().sum()) / len(df))
    nanval = df.isnull().sum()
    
    #creating a table 
    
    nanvaltable = pd.concat([nanval,nanval_perct], axis =1)
    nanvaltable = nanvaltable.rename(columns = \
                        {0 : 'Sum_MissingValues', 1 : 'Perc_MissingValues'  })
   
    #Now that we got a table with this information, 
    #I am going to proceed to extract only the variables with null values
    
    filtered_null_data = nanvaltable[nanvaltable.Perc_MissingValues !=0].sort_values(\
                                                        'Perc_MissingValues', ascending=False).round({'Perc_MissingValues':3})
    
    return filtered_null_data
```

```python
# converting emp_length to float
dict_emp_length = {'10+ years':10, '6 years':6, '4 years':4, '< 1 year':0.5, '2 years':2,
       '9 years':9, '0 years':0, '5 years':5, '3 years':3, '7 years':"7", '1 year':1,
       '8 years':8}

def emp_lengt_tonum (emp_length):
    if emp_length == '10+ years':
        return  '10yrs'
    elif emp_length == '6 years':
        return 'btw 4 - 6yrs'
    elif emp_length == '4 years':
        return 'btw 4 - 6yrs'
    elif emp_length == '< 1 year':
        return '< 4yrs'
    elif emp_length == '2 years':
        return '< 4yrs'
    elif emp_length == '9 years':
        return 'btw 7 - 9yrs'
    elif emp_length == '0 years':
        return '< 4yrs'
    elif emp_length == '5 years':
        return 'btw 4 - 6yrs'
    elif emp_length == '3 years':
        return '< 4yrs'
    elif emp_length == '7 years':
        return 'btw 7 - 9yrs'
    elif emp_length == '1 year':
        return '< 4yrs'
    elif emp_length == '8 years':
        return 'btw 7 - 9yrs'
    else:
        return emp_length

    

df_c['emp_length'] = df_c['emp_length'].apply(emp_lengt_tonum)
```

```python
Q25 = df_c['loan_amnt'].quantile(0.25)
Q50 = df_c['loan_amnt'].quantile(0.50)
Q75 = df_c['loan_amnt'].quantile(0.75)
print(Q25, Q50, Q75)
```

```python
def loan_amounttype(loan_amnt):
    if loan_amnt <= Q25:
        return 'Small(<=8000)'
    if (loan_amnt > Q25) & (loan_amnt <= Q50):
        return 'Regular(>8000 & <=12900)'
    elif (loan_amnt > Q50) & (loan_amnt <= Q75):
        return 'Medium(>12900 & <=20000)'
    else:
        return 'Large(>20000)'
    
#======================================

df_c['loan_amnty'] = df_c[['loan_amnt']].applymap(lambda x : loan_amounttype(x))
df_c['loan_amnty'].value_counts()
```

```python

def counting_nullval (df):
    null_perct = ((100 * df.isnull().sum()) / len(df))
    null_sum = df.isnull().sum()
    
    #creating a table 
    
    nullvaltable = pd.concat([null_sum,null_perct], axis =1)
    nullvaltable = nullvaltable.rename(columns = \
                        {0 : 'Sum_MissingValues', 1 : 'Perc_MissingValues'  })
    
    filtered_null_data = nullvaltable[nullvaltable.Perc_MissingValues !=0].sort_values(\
                                                        'Perc_MissingValues', ascending=False).round({'Perc_MissingValues':3})
    
    return filtered_null_data

#======================================

nul_values = counting_nullval(df_lcf)
nul_values.head(10)
```

The main metric to consider for compare and validate the models will be ROC-AUC score. The reason is beacuse on this model, I care about the default samples, the class inbalance. The reazon of choosing the ROC-AUC score is because it is independent of the threshold set for classification and because it only considers the rank of each prediction and not its absolute value.

Other metrics I am checking in the models are:


    
F1 Score: It is a harmonic mean of precision and recall given by- F1 = 2*Precision*Recall/(Precision + Recall)
This is beause will provide the model performance.
